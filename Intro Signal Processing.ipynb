# 1D Signals 
### Generally all functions have the following layout:

$y(t)= sin(nπft)$ where n = 2, f is the function frequency, and t is time

```python
def gen_wave(func,f,t):
    print('f=',f,'periods/(2π*100)samples | Freq= {:.4f} period/s'.format(f/(2*np.pi)))
    return func(f*t)


plot.rcParams.update(params)

plt.figure(figsize=(20,5))
time = np.arange(0,2*np.pi,1/100) 
plt.axhline(y=0.0, color='c', linestyle='--')
plt.plot(gen_wave(func=np.sin,f=1,t=time),label='f=1')
plt.xlabel("Sample #")
plt.show()
plt.figure(figsize=(20,10)) 
plt.axhline(y=0.0, color='c', linestyle='--')
plt.plot(time,gen_wave(func=np.sin,f=1,t=time),label='f=1')
plt.plot(time,gen_wave(func=np.sin,f=2,t=time),label= 'f=2')
plt.plot(time,gen_wave(func=np.sin,f=5,t=time),label='f=5')
plt.xlabel('seconds')
plt.legend()
plt.show()
```

For all signals y:

If y is periodic, there is a `period (T)` = `1/sampling_freq` <br />
where `sampling_freq (f)`= #samples/second = `1/T`
<br />

Generally signal equations are governed by $y(t) = func(nπft)$  <br />


In the case below we examine signal where $y(t)= sin(x) + cos(2x)$ <br />
Each sample is examined at `f_s= 100 samples/sec` <br />
- sin(x) --> $x =2πfx$ --> frequency of function = 1/2π ~ .159$\frac{1}{s}$  ..... period ~ 6.28s
- cos(2x) --> $2x =2πfx$ --> frequency of function = 1/π ~ .318$\frac{1}{s}$ ..... period ~ 3.14s



```python
s_f=100 #samps/sec
T=1/s_f
time = np.arange(0,2*np.pi,T)   # start,stop,step

plt.figure(figsize=(20,5))
plt.suptitle('Examination of Signal Isolation and Construction')
plt.subplot(131)
y_sin2x = np.sin(2*time) 
plt.plot(time,y_sin2x,color='b')
plt.xlabel("time [s]")
plt.title("sin(2x) | Period=π")
plt.subplot(132)
y_cos4x = np.cos(4*time) 
plt.plot(time,y_cos4x,color='r')
plt.xlabel("time [s]")
plt.title("cos(4x) | Period= π/2")
plt.subplot(133)
plt.title("sin(x) + cos(2x) | Period= π")
plt.plot(time,y_sin2x,color='b',label='sin(x)')
plt.plot(time,y_cos4x,color='r',label='cos(2x)')
plt.xlabel("time [s]")
plt.legend(loc='upper right')
plt.show()
```

### Examining superimposed signal 

```python
def gen_func(s_f=100,plot=False):
    T=1/s_f
    x = np.arange(0,2*np.pi,T)   # start,stop,step
    print('Sample_freq=', s_f,'samps/sec')
    y = np.sin(2*x) + np.cos(4*x) 
    if plot == True:
        plt.plot(x,y,color='b')
        plt.title('Sample Signal w/ Period Len = {:.2f} samps'.format(np.pi))
        plt.axhline(y=0.0, color='c', linestyle='--')
        min_val= x[np.where(y == min(y[np.where(x<1.2*np.pi)]))]
        for i in range(2):
            plt.axvline(x=min_val+np.pi*i, color='r', linestyle='--')
        plt.xlabel("[0 --- 2π]")
        return None
    else:
        return x,y
    
gen_func(s_f=100,plot=True)
```

### Generating Square Wave Function

$y(t)= \frac{4}{π}*\sum_{n=1,3,5...}^{\infty }\frac{1}{n}sin(\frac{nπt}{T_L}))$



```python
def gen_square_func(num_plots=20,n=8,s_f=100):
    T= 1/s_f
    y = 0
    for i in range(1,num_plots,2):
        y += 1/i*np.sin(i*2*time) 
    y=y*4/np.pi
    plt.plot(time,y,color='m')
    plt.axhline(y=0.0, color='c', linestyle='--')
    plt.xlabel("Sample #")
    
gen_square_func(num_plots=100,n=10)
gen_func(s_f=100,plot=True)
```



# Fourier transforms shift from the physical realm -->  frequency domain 
---
## Solving for $a_n$

### $\int_{-T_L}^{T_L} cos(mπft)*y(t)dt = \int_{-T_L}^{T_L} \frac{1}{2}*a_0*cos(mπft)+ \sum_{n=0}^{N-1}*cos(mπft)*(a_n*cos(\frac{nπt}{T_L})+b_n*sin(\frac{nπt}{T_L}))dt$

### > $\int_{-T_L}^{T_L} \frac{1}{2}*a_0*cos(mπft)dt + \sum_{n=0}^{N-1}\int_{-T_L}^{T_L}cos(mπft)*a_n*cos(\frac{nπt}{T_L})dt +\int_{-T_L}^{T_L}*cos(mπft)*b_n*sin(\frac{nπt}{T_L})dt$

### > $\int_{-T_L}^{T_L} \sum_{n=0}^{N-1}cos(mπft)*a_n*cos(\frac{nπt}{T_L})dt = a_n*T_L\text{ if m=n, otherwise 0.}$

### > $a_n= \frac{1}{T_L} * \int_{-T_L}^{T_L} cos(mπft)*y(t)dt$


---
## Solving for $b_n$

### $\int_{-T_L}^{T_L} sin(mπft)*y(t)dt = \int_{-T_L}^{T_L} \frac{1}{2}*a_0*sin(mπft)+ \sum_{n=0}^{N-1}*sin(mπft)*(a_n*cos(\frac{nπt}{T_L})+b_n*sin(\frac{nπt}{T_L}))dt$

### > $\int_{-T_L}^{T_L} \frac{1}{2}*a_0*sin(mπft)dt + \sum_{n=0}^{N-1}\int_{-T_L}^{T_L}sin(mπft)*a_n*cos(\frac{nπt}{T_L})dt +\int_{-T_L}^{T_L}*sin(mπft)*b_n*sin(\frac{nπt}{T_L})dt$

### > $\int_{-T_L}^{T_L} \sum_{n=0}^{N-1}sin(mπft)*b_n*sin(\frac{nπt}{T_L})dt = b_n*T_L \text{ if m=n, otherwise 0.}$

### > $b_n= \frac{1}{T_L} * \int_{-T_L}^{T_L} sin(mπft)*y(t)dt$

---
## Reducing $a_n + b_nj$

### $a_n + b_nj= \frac{1}{T_L} * \int_{-T_L}^{T_L} cos(mπft)*y(t)dt + j*sin(mπft)*y(t)dt$

### > $e^{mπft} = cos(2πft)+i*sin(mπft) $

### > $a_n + b_nj= \frac{1}{T_L} * \int_{-T_L}^{T_L} y(t)*e^{imπft}dt$

### > $a_n + b_nj= \int_{-T_L}^{T_L} y(t)*e^{-i2πft}dt$


******************
### $e^{-i2πft} = e^{\frac{−i 2πkn}{N}} \text{, N= number of points in the signal, k && n= array len of signal}$ 
#### - $\text{frequency = number points in signal / period length}$
#### - $f = \frac{k*s_f}{N}$
#### - $t = \frac{n}{s_f}$
#### - $f*t= \frac{k*n}{N}$
*******************

###  $\vec{a_n}+ \vec{b_n}j = M*\vec{y} \text{  , where }M = e^{\frac{−i 2πkn}{N}}$

```python
def frequency_range(y,s_f=100):
    """
    Determines a signals frequency range 
    """
    N = len(y) # length of the signal
    k = np.arange(N)
    T = N/s_f
    return k/T
def time_range(y,s_f=100):
    """
    Determines a signals time range 
    """
    return np.arange(len(y))/s_f
```

```python
s_f=100
def plot_fft(y):
    frq=frequency_range(y)[0:int(len(y)/2)] # one sided frequency range
    c= abs(np.fft.fft(y))
    c=c[0:int(len(y)/2)]
    plt.plot(frq,c) # fft computing and normalization
    plt.xlim(0, s_f / 40) # Zoom in 
    plt.xlabel('Frequencies')
    max_loc=c.argsort()[-2:]
    for loc in max_loc:
        plt.axvline(x=frq[loc], color='r', linestyle='--')
    print('Max frequencies are 1/π ({:.4f}) and 2/π ({:.4f})'.format(frq[max_loc][0],frq[max_loc][1]))
plot_fft(y)
```

## $\vec{Y} = M*\vec{y} \text{  , where }M = e^{\frac{−i 2π k n} {N}} = e^{-i2πft}$

N= len of signal 
n = range [0 N]
k = [N x 1] array
M = N X N of exp(π/N*k*n)

```python
s_f=100

def DFT_vectorized(x):
    """Compute the discrete Fourier Transform of the 1D array x"""
    N = int(x.shape[0])
    n = np.arange(N)
    k = n.reshape(N, 1)
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)[0:int(len(x)/2)]

def DFT_vectorized_time_frq(x,time,freq):
    """
    Compute the discrete Fourier Transform of the 1D array x.
    The following proves time and freq can be estimated 
    """
    M = np.exp(-2j * np.pi * time * np.expand_dims(freq,axis=1))
    return np.dot(M, x)[0:int(len(x)/2)]

frq=frequency_range(y)[0:int(len(y)/2)]
plt.figure(figsize=(20,4))
plt.subplot(121)
plt.title('DFT w time and freq range information')
plt.plot(frq,abs(DFT_vectorized_time_frq(y,time,frq)))
plt.xlim(0, s_f / 10) # Zoom in 
plt.xlabel('Frequencies')
plt.subplot(122)
plt.title('DFT w/o time and freq range information')
plt.plot(frq,abs(DFT_vectorized(y)))
plt.xlim(0, s_f / 10) # Zoom in 
plt.xlabel('Frequencies')
plt.show()
```
